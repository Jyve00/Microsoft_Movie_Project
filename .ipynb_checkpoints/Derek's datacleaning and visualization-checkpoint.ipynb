```python
tn = pd.read_csv('tn.movie_budgets.csv')
```

```python

tn
```

```python
tn['production_budget'] = tn['production_budget'].replace(',','', regex = True)
tn['production_budget'] = tn['production_budget'].str.replace('$','', regex = True)
tn['domestic_gross'] = tn['domestic_gross'].replace(',','', regex = True)
tn['domestic_gross'] = tn['domestic_gross'].str.replace('$','', regex = True)
tn['worldwide_gross'] = tn['worldwide_gross'].replace(',','', regex = True)
tn['worldwide_gross'] = tn['worldwide_gross'].str.replace('$','', regex = True)

#replacing commas and dollarsigns to turn str to int
```

```python
tn['production_budget']= tn['production_budget'].astype(float)
tn['domestic_gross']= tn['domestic_gross'].astype(float)
tn['worldwide_gross']= tn['worldwide_gross'].astype(float)
```

```python
tn['profit']= tn['worldwide_gross']-tn['production_budget']
tn['ROI']= (tn['profit']/tn['production_budget'])*100
#creating new columns for profit and roi
```

```python
tn = tn.drop('release_date', 1)
tn = tn.drop('id', 1)
#dropping release date and id
```

```python
tn = tn.rename({'movie':'title'}, axis = 1)
```

```python
final_merge = pd.merge(final_merge, tn, on ='title')
```

```python
final_merge
```

# data visualization 



```python
directors = final_merge[final_merge['job'] == 'director']
```

```python
unique_titles_df = directors.drop_duplicates(subset= ['title'])
```

```python
final_merge = final_merge.drop('domestic_gross_x', 1)
final_merge = final_merge.drop('worldwide_gross_x', 1)
```

```python
directors.info()
```

```python
profit_directors = directors.groupby('primary_name')['profit'].sum().sort_values(ascending=False)

```

```python
top_50_directors = profit_directors.nlargest(50)
```

```python
top_50_directors_profit = top_50_directors.values.sum()

#created a top_50 directors based on profit
```

```python
total_profit = unique_titles_df['profit'].sum()
```

```python
top_50_directors_profit_perc = top_50_directors_profit/total_profit

#dividing the total profit from the top 50 directors by the total profit in the dataset to see how much can be attributed to top 50 directors
```

```python
top_50_directors_profit_perc
```

```python
unique_titles_df[unique_titles_df['title'].str.contains('Avengers')]
```

```python
rest_profit = 1 - top_50_directors_profit_perc 
```

```python
import matplotlib.pyplot as plt
```

```python
my_data = [rest_profit, top_50_directors_profit_perc]
my_labels = 'Everyone else', 'Top 50 profitable directors'
my_colors = ['lightblue','silver']
my_explode = (0, 0.1)
plt.pie(my_data,labels = my_labels,autopct='%1.1f%%', startangle=15, shadow = True, colors=my_colors, explode=my_explode)
plt.title('Total Profit')
plt.axis('equal')
plt.show()

#graphing pie plot to show signifigance
```

```python
import pandas as pd
import numpy as np
```

```python
m_b = pd.read_csv('Data/tn.movie_budgets.csv')
m_b
```

```python
m_b = m_b.drop(columns=['release_date', 'id', 'domestic_gross'], axis = 1)
```

```python
m_b.rename({'movie' : 'title'}, axis = 1, inplace=True)
```

```python
m_b['production_budget'] = m_b['production_budget'].str.replace('$', '')
m_b['production_budget'] = m_b['production_budget'].str.replace(',', '')
m_b['worldwide_gross'] = m_b['worldwide_gross'].str.replace('$', '')
m_b['worldwide_gross'] = m_b['worldwide_gross'].str.replace(',', '')
```

```python
m_b
```

```python
m_b['profit'] = m_b['worldwide_gross'].astype('int64') - m_b['production_budget'].astype('int64')
```

```python
m_b
```

```python
m_b = m_b.sort_values(by=['profit'], ascending=False)
```
